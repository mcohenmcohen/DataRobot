```python
import pandas as pd
import numpy as np
import datetime
```

```python
todays_date = datetime.datetime.now().date()
index = pd.date_range(todays_date, periods=10, freq='D')
cols = ['Sensor_1','Sensor_2','Sensor_3']
df = pd.DataFrame(np.random.randint(5, size=(10, 3)), index=index, columns=cols) 
df
```

```python
for col in cols:
    df[col+'_t-1'] = df[col].shift(1)
    df[col+'_t-2'] = df[col].shift(2)
    df[col+'_avg3'] = df[col].rolling(3).mean()
    df[col+'_std3'] = df[col].rolling(3).std()
df
```
