```python
soup = bs4.BeautifulSoup(html, 'lxml')
```

```python
leaderboard = soup.find(name='div', attrs={'id':'leaderBoard-data-scroll'})
```

```python
models = leaderboard.find_all(name='div', attrs={'class':'lb-row'})
```

```python
columns = ['name', 'desc', 'features', 'sample_size', 'score', 'cv_score', 'holdout_score']
res = pd.DataFrame(columns=columns)

for i,model in enumerate(models):
    print(i)
    row = pd.Series(index=columns)
    row['name'] = model.find(name='div', attrs={'class':'dr-cell column-name'}) \
                       .find(name='span', attrs={'class':'model-name'}).text.strip()
    if model.find(name='div', attrs={'class':'dr-cell column-name'}) \
            .find(name='span', attrs={'class':'features ng-binding ng-scope'}):
        row['desc'] = model.find(name='div', attrs={'class':'dr-cell column-name'}) \
                           .find(name='span', attrs={'class':'features ng-binding ng-scope'}).text.strip()
    row['features'] = model.find(name='div', attrs={'class':'dr-cell col-model-retrain'}) \
                           .find(name='span', attrs={'class':'feature-list cell-text ellipse-content'}).text.strip()
    row['sample_size'] = model.find(name='div', attrs={'class':'dr-cell col-model-retrain'}) \
                              .find(name='span', attrs={'test-id':'sample-size-label'}).text.strip()
    row['score'] = model.find(name='div', attrs={'class':'dr-cell col-score'}) \
                        .find(name='format-data').text.strip()
    if model.find(name='div', attrs={'class':'dr-cell col-score cv-column-cell ng-scope'}) \
            .find(name='format-data'):
        row['cv_score'] = model.find(name='div', attrs={'class':'dr-cell col-score cv-column-cell ng-scope'}) \
                               .find(name='format-data').text.strip()
    row['holdout_score'] = model.find(name='div', attrs={'class':'dr-cell col-score holdout-column-cell ng-scope'}) \
                                .find(name='format-data').text.strip()
    res = res.append(pd.DataFrame(row).T)
```
