```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'date':pd.date_range(start='1/1/2018', end='1/10/2018'),
    'numbers': range(1,11)
})
df
```

```python
df['lag 1'] = df.numbers.shift(1)
df['lag 2'] = df.numbers.shift(2)
df['lag -1 (future leak!)'] = df.numbers.shift(-1)
df['5 day rolling mean'] = df.numbers.rolling(5).mean()
df['5 day rolling standard deviation'] = df.numbers.rolling(5).std()
df['1 day pct change'] = df.numbers.pct_change()
df['5 day pct change'] = df.numbers.pct_change(5)
df
```
