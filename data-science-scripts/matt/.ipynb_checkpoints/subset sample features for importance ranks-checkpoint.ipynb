## Create a factory of models, each model trained using a subset of features.  Then aggregate the top features, ranking by their feature importance across all models, and presenting in box plot form.

## This is helpful when you have more features than rows, or if you want a deeper analysis on data with very large numbers of features.

For each model (and you should just pick one, probably the best model in the project, like an xgboost):
- randomly select a subset of features (say, 100)
- create a feature list of the selected features
- train the model on the feature list
- generate the feature impact
- capture the rank order of the top features from feature impact (say 25), 1 being top ranked

When all models and feature impacts are calcultated, for each feature:
- aggregate the ranks across all mmodels
- average the ranks
- display as box plots
- *the lowest average ranks are the top features

```python
import pandas as pd
import numpy as np
import os
import sys
import time
import datarobot as dr
import datarobot.errors as dre
import requests.exceptions as rex
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.options.display.max_columns=500
pd.options.display.max_info_rows=500
pd.options.display.max_rows=500
pd.options.display.max_colwidth=100

dr.Client(token=os.environ['DATAROBOT_API_TOKEN'], endpoint='https://app.datarobot.com/api/v2')

# Random sampling parameters:
# random_sample_size - number of features to randomly sample to build a feature list with
# num_top_features - the top number of features to get from a completed model's feature impact
# num_sample_runs - number of feature sampled models to run
RANDOM_SAMPLE_SIZE=100
NUM_TOP_FEATURES=100
NUM_SAMPLE_RUNS=200
```

```python
def subsample_feature_importance(project, random_sample_size=100, 
                                          num_top_features=100, 
                                          num_sample_runs=3,
                                          train_sample_pct=20):
    '''
    Ramdom sample a subset of features
    Create a feature list
    Train a model of the new featurelist
    Get the model's feature importance 
    Capture the rank of the important features
    Aggreagate all feature ranks across all trained models
    
    Parameters
        project - DataRobot project object
        random_sample_size - number of features to random sample to build a feature list with
        num_top_features - the top number of features to get from a model's feature impact
        num_sample_runs - number of sample runs to perform
    
    Return 
        Datafrane of feature ranks.
    '''
    # Time this
    tt = time.time()
    
    print('Start subsample_feature_importance')
    print('- random_sample_size: %s' % random_sample_size)
    print('- num_top_features: %s' % num_top_features)
    print('- num_sample_runs: %s' % num_sample_runs)
    print()
    
    # Set worker count to max for user's account
    def get_worker_limit(project):
        try:
            project.set_worker_count(999999)
        except dre.ClientError as ce:
            limit = ce.json.get('message').rsplit(None, 1)[-1].replace(')', '')
        return limit
    limit = get_worker_limit(project)
    print('Set max worker count:', limit)
    project.set_worker_count(int(limit))

    # Get the project features
    features = np.array([feat.name for feat in project.get_features() if feat.name != project.target])
    print('Project \'%s\' has %s features' % (project.id, len(features)))

    # Get top model from the leaderboard to do our sampling
    models = project.get_models()
    for model in models:
        if 'Blender' not in model.model_type:
            best_model = model
            break
    model = best_model
    print('Best model from the leaderboard: \'%s\'' % model.model_type)
    print()

    # Create a feature list of n randomly drawn features, and 
    # Queue a model job to be trained with it
    model_job_ids = []  # no longer used
    start_idx = 1
    do = True
    t1 = time.time()
    while do:
        for i in range(start_idx, num_sample_runs+1):
            # Client, async process, or timeout errors may happen
            try:
                # sample_features and create feature list
                sample_features = np.random.choice(features, random_sample_size, replace=False).tolist()
                featurelist_name_prefix = 'random_sample_feature_list_'
                featurelist_name = featurelist_name_prefix + str(i)
                sample_featureslist = project.create_featurelist(featurelist_name, sample_features)

                # Train model on feasture list
                print('Request model training with feature list \'%s\'' % featurelist_name, end="\r", flush=True)
                model_job_id= model.train(sample_pct=train_sample_pct,
                                          featurelist_id=sample_featureslist.id,
                                          scoring_type=dr.SCORING_TYPE.validation)
                model_job_ids.append(model_job_id)
                if i == num_sample_runs:
                    do = False
                    break
            except dre.ClientError as ce:
                # A datarobot API exception
                print('\n- ClientError - %s' % ce)
                print('- Exiting...')
                return
            except Exception as e:
                # This exception is likely a timeout.  Output the message, reset the loop index and continue.
                # We still want to capture the loops performed and return the dataframe.
                start_idx = i
                print('\n- Exception:', e)
                print('- Exiting...')
                return
    print('\n- All models queued to train with a randomly generated feature list: %0.3f minutes' % \
          (float((time.time()-t1)/60)))
    print()
                
    # Wait for remaining model training jobs to complete
    t1 = time.time()
    while True:
        # gives all jobs queued or inprogress
        fi_jobs = [mj for mj in project.get_all_jobs() if mj.job_type == 'model']  
        print('Wait for all model training jobs to complete. %s models remaining   ' % \
              len(fi_jobs), end="\r", flush=True)
        if len(fi_jobs) == 0:
            break
        time.sleep(1)
    print('\n- All model training jobs complete: %0.3f minutes' % (float((time.time()-t1)/60)))


# Run models with sampled features
subsample_feature_importance(project, 
                             random_sample_size=RANDOM_SAMPLE_SIZE, 
                             num_top_features=NUM_TOP_FEATURES, 
                             num_sample_runs=NUM_SAMPLE_RUNS)
```

```python
def get_feature_impacts(project, num_top_features=100):
    
    # Time this
    tt = time.time()
    
    models = project.get_models()
    
    # Request model feature impact jobs
    t1 = time.time()
    for i, m in enumerate(models):
        print('model %s: %s' % (i, (m.model_type, m.model_number, m.id)))
        try:
            fi_job = m.request_feature_impact()
            print('Requesting model feature impact jobs. %s of %s models remaining   ' % \
                  (str(len(models)-i-i), i), end="\r", flush=True)
        except dre.JobAlreadyRequested as jar:
            # There should be no previously requested fi job, 
            # but if there are that's ok, move on to the next request
            print('- JobAlreadyRequested error:', jar)
        except Exception as e:
            # There should be no previously requested fi job, 
            # but if there are that's ok, move on to the next request
            print('- Exception error:', e)
    print('\n- Feature impact for all models requested: %0.3f minutes' % (float((time.time()-t1)/60)))
    print()
    
    # Wait for model feature impact jobs
    t1 = time.time()
    while True:
        # gives all jobs queued or inprogress
        fi_jobs = [mj for mj in project.get_all_jobs() if mj.job_type == 'featureImpact']  
        print('Wait for all feature impact jobs to complete. %s jobs remaining   ' % \
              len(fi_jobs), end="\r", flush=True)
        if len(fi_jobs) == 0:
            break
        time.sleep(1)
    print('\n- All feature impact jobs complete: %0.3f minutes' % (float((time.time()-t1)/60)))

    # Get model feature impact jobs
    t1 = time.time()
    model_top_features_dict = {}
    for i, m in enumerate(models):

        print('Retrieving model feature impact. %s remaining of %s models   ' % \
              (str(len(models)-i-1), i), end="\r", flush=True)

        feature_impacts = m.get_feature_impact()

        feature_impacts.sort(key=lambda x: x['impactNormalized'], reverse=True)
        final_names = [f['featureName'] for f in feature_impacts[:num_top_features]]
        final_names

        # Store the model's top features in a dict - model.id : list(top_features)
        model_top_features_dict[m.id] = final_names
    print('\n- Feature impact for all models retrieved: %0.3f minutes' % (float((time.time()-t1)/60)))
    print()

    # Aggregate the rank order for each top n feature in each subsample model
    print('Building feature ranks dataframe')
    print()
    feature_rank_dict = {}
    for mkey in model_top_features_dict.keys():
        top_feats = model_top_features_dict.get(mkey)
        for i, feat in enumerate(top_feats):
            if feat in feature_rank_dict:
                feat_list = feature_rank_dict[feat]
                feat_list.append(i+1)
            else:
                feature_rank_dict[feat] = [i+1]

    # Put top n features of each model into a dataframe along with some summary statistics
    df = pd.DataFrame(data={'feature': list(feature_rank_dict.keys()), 'ranks': list(feature_rank_dict.values())})
    df['ranks'] = df['ranks'].map(lambda x: str(x).lstrip('[').rstrip(']'))
    df['mean'] = df['ranks'].map(lambda x: np.array(x.split(',')).astype(np.float).mean())
    df['median'] = df['ranks'].map(lambda x: np.median(np.array(x.split(',')).astype(np.float)))
    df['count'] = df['ranks'].map(lambda x: np.array(x.split(',')).shape[0])
    cols = ['feature', 'ranks', 'mean', 'median', 'count']
    df = df[cols]
    df = df.sort_values(['count', 'median'], ascending=[False, True])

    print('*** Total Time: %.3f minutes' % (float((time.time()-tt)/60)))

    return df
    
df_feat_imps = get_feature_impacts(project, num_top_features=NUM_TOP_FEATURES)
```

```python
def delete_user_featurelists(project):
    '''
    Delete user features lists from a project built for random sampling.
    Namely, from a previous random sample runs that didn't complete.
    '''
    t1 = time.time()
    while True:
        try:
            featurelists = [fl for fl in project.get_featurelists() if 'random_sample_feature_list_' in fl.name]
            fl_remain = len(featurelists)
            if len(featurelists) == 0:
                break
            for fl in featurelists:
                fl_remain -= 1
                print ('Deleting %s.  %s feature lists remaining.' % (fl, fl_remain), end="\r", flush=True)
                fl.delete(delete_dependencies=True)
        except:
            # If there is an exception, output the message and proceed.
            # We still want to capture the loops performed and return the dataframe.
            print('- Exception:', sys.exc_info()[0])
            break
    print('\n- All feature lists deleted: %0.3f minutes' % (float((time.time()-t1)/60)))

# Delete the random sample features lists and associated models for clean-up
# delete_user_featurelists(project)
```

```python
def feature_boxplots(in_df):
    '''
    Plot a seaborn boxplot for all sampled feature ranks from model feature impacts
    '''
    df_feat_imps = in_df.copy()
    
    # Remove any features not sampled at least a threshold amount.  Greater than 1/3 max count, and at least three.
    thresh = max(df_feat_imps['count'].max() / 3, 3)
    df_feat_imps = df_feat_imps[df_feat_imps['count'] > thresh]
    
    dfe = pd.DataFrame()

    for idx in range(df_feat_imps.shape[0]):
        feat = df_feat_imps.iloc[idx]['feature']
        ranks = [s.strip() for s in df_feat_imps.iloc[idx]['ranks'].split(',')]
        df_col = pd.DataFrame()

        df_col[feat] = ranks
        df_col[feat] = df_col[feat].astype(int)

        dfe = pd.concat([dfe, df_col], axis=1)

    columns = dfe.columns.tolist()
    dfe = dfe.reindex(columns=columns)
    dfe[columns] = dfe[columns].astype(float)

    meds = dfe.median().sort_values()
    dfe = dfe[meds.index]

    fig, ax = plt.subplots(figsize=(15,int(df_feat_imps.shape[0]*.3)))
    sns.boxplot(data=dfe[meds.index], orient='h', ax=ax)
    
print('Num df_feat_imps:', df_feat_imps.shape)
print('Minimum times a feasture must have appeareed across all feature sampling models in order to count:')
print('- %s' % max(df_feat_imps['count'].max() / 3, 3))
print()

# Plot the box plots of the important features
if df_feat_imps is not None:
    print('Generate boxplots...')
    feature_boxplots(df_feat_imps)
```
