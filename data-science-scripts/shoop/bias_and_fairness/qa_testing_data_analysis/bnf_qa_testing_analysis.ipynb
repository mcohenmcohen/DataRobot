```python
import json

import pandas as pd
```

```python
cross_class_accuracy_results = []
cross_class_disparity_results = []

with open("model_cross_class_accuracy_ted.json", "rb") as f:
    for line in f:
        cross_class_accuracy_results.append(json.loads(line))

with open("model_cross_class_disparity_ted.json", "rb") as f:
    for line in f:
        cross_class_disparity_results.append(json.loads(line))
```

```python
print(f"Cross class accuracy results length: {len(cross_class_accuracy_results)}")
print(f"Cross class data disparity results length: {len(cross_class_disparity_results)}")
```

```python
df1 = pd.DataFrame(
    columns=["model_id", "feature", "class_name", "LogLoss", "f1", "AUC", "accuracy", "prediction_threshold"]
)
df2 = pd.DataFrame(
    columns=[
        "model_id", "feature", "class1_name", "class1_total", "class2_name",
        "class2_total", "selected_feature", "disparity_score", "feature_impact"
    ]
)
```

```python
for cross_accuracy in cross_class_accuracy_results:
    row = {}
    row["model_id"] = cross_accuracy["model_id"].get("$oid")
    row["feature"] = cross_accuracy["feature"]
    row["prediction_threshold"] = cross_accuracy["prediction_threshold"]
    for class_segment in cross_accuracy["per_class_accuracy_scores"]:
        row["class_name"] = class_segment["class_name"]
        metrics_scores = {
            score["metric"]: score["value"]
            for score in class_segment["metrics"]
        }
        row["LogLoss"] = metrics_scores["LogLoss"]
        row["f1"] = metrics_scores["f1"]
        row["AUC"] = metrics_scores["AUC"]
        row["accuracy"] = metrics_scores["accuracy"]
        # save results
        df1 = df1.append(row, ignore_index=True)
```

```python
for cross_disparity in cross_class_disparity_results:
    row = {}
    class1, class2 = cross_disparity["class_labels"][0], cross_disparity["class_labels"][1]
    
    row["model_id"] = cross_disparity["model_id"].get("$oid")
    row["feature"] = cross_accuracy["feature"]
    row["class1_name"] = class1
    row["class2_name"] = class2
    row["class1_total"] = cross_disparity["data"][0][f"{class1}_total"]
    row["class2_total"] = cross_disparity["data"][0][f"{class2}_total"]
    for selected_feature in cross_disparity["data"]:
        row["selected_feature"] = selected_feature["name"]
        row["disparity_score"] = selected_feature["disparity_score"]
        row["feature_impact"] = selected_feature["feature_impact"]
        df2 = df2.append(row, ignore_index=True)
```
